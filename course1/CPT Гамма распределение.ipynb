Приближаем нормальным распределением с параметрами (EX, DX/n), где EX и DX матожидание и дисперсия гамма-распределения   (вычислены заранее)   
EX = mean     
DX/N = var/N    

```python
#функция построения гистрограмм распределения выборочных средних
def generate_pdf(n): 
    #генер. выбороч. средние 
    means = []
    for _ in range(1000):
        sample = rv.rvs(n)
        sample_mean = np.mean(sample)
        means.append(sample_mean)
    
    #рассчет теор. плотности вероятности для нормального распределения 
    mu = mean 
    sigma = (var/n) ** 0.5
    norm_rv = sts.norm(loc=mu, scale=sigma)
    x = np.linspace(0,4,100)
    pdf = norm_rv.pdf(x)
    
    #строим гистограмму выборочных средних и оценку плотности вероятности норм. расп. 
    plt.plot(x, pdf)
    plt.hist(means, density = True)
    plt.xlabel('$\\bar{X}_n$')
    plt.ylabel('$f(\\bar{X}_n)$')
    plt.title('n = %d' % n)
    plt.show()
```

```python
#строим для выборок объема 5 10 50 гистограммы расп. выборочных сердних и приблежения плотностями норм. расп. 
for n in [5, 10, 50]:
    generate_pdf(n)
```
